# EDA — Commentaires libres & coût support

## Objectif
Analyser les commentaires libres des tickets support afin d’évaluer :
- leur exploitabilité pour identifier des causes racines
- leur lien avec le coût opérationnel (temps de résolution)

## Données utilisées
- data/raw/support_tickets.csv
- data/raw/clients.csv
- data/raw/products.csv

## Démarche
1. Audit de la qualité du texte
2. Analyse du coût support (TTR)
3. Identification de thèmes dominants
